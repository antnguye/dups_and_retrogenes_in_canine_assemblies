```python
# build a new overlap level - collapse all matches
mischka_biser_overlap = {}
mischka_biser_unique = []
total_biser_dups = 0
biser_remove_1st_mb = 0
f = open("mischka.biser_to_fastCN_union.bed", 'rt')
for line in f:
    line = line.rstrip().split()
    # remove unplaced
    if "chrUn" in line[0] or "chrY" in line[0] or "chrM" in line[0]:
        continue
    total_biser_dups += 1
        
    size = int(line[2]) - int(line[1])
    if int(line[1]) < 1000000: #check 1st mb
        continue
    elif size < 10000: #check 10kb
        continue
        

    
    biser_dup = (line[0], line[1], line[2])
    
    if line[3] == ".": # no fastCN overlap
        biser_remove_1st_mb += 1
        biser_dup = (line[0], line[1], line[2], size)
        mischka_biser_unique.append(biser_dup)
    elif int(line[4]) < 1000000 or (int(line[5]) - int(line[4]) < 10000): # check for 1stmb and size for fastCN
        continue
    else: # overlap found
        biser_remove_1st_mb += 1
        overlap = int(line[7])
        unique = size - overlap
        if biser_dup not in mischka_biser_overlap.keys():
            mischka_biser_overlap[biser_dup] = []
        mischka_biser_overlap[biser_dup].append((overlap, unique, size))
    
f.close()
mischka_biser_overlap
```

```python
msk_chro_sizes = {}
inFile = open("mischka.sorted.chrom.sizes", 'rt')
for line in inFile:
    line = line.rstrip().split()
    if "chrUn" in line[0] or "chrY" in line[0] or "chrM" in line[0]:
        continue
    msk_chro_sizes[line[0]] = int(line[1])
inFile.close()
msk_chro_sizes
```

```python
msk_all_biser = {}
msk_biser_support = {}
msk_biser_no_support = {}
inFile = open('mischka.biser_to_fastCN_union.bed', 'rt')
for line in inFile:
    line = line.rstrip().split()
    if "chrUn" in line[0] or "chrY" in line[0] or "chrM" in line[0]:
        continue
    chro = line[0]
    size = int(line[2]) - int(line[1])
    

    if int(line[1]) < 1000000: #check 1st mb
        continue
    elif size < 10000: #check 10kb
        continue
    elif int(line[4]) < 1000000 or (int(line[5]) - int(line[4]) < 10000): # check for 1stmb and size for fastCN
        continue
    
    if chro not in msk_all_biser.keys():
        msk_all_biser[chro] = 0
    msk_all_biser[chro] += size
    
    if line[3] == ".": #no fastCN support
        if chro not in msk_biser_no_support.keys():
            msk_biser_no_support[chro] = 0
        msk_biser_no_support[chro] += size
    else:
        try:
            support = int(line[7])
        except:
            print(line)
        not_supported = size - support
        
        if chro not in msk_biser_no_support.keys():
            msk_biser_no_support[chro] = 0
        msk_biser_no_support[chro] += not_supported
        
        if chro not in msk_biser_support.keys():
            msk_biser_support[chro] = 0
        msk_biser_support[chro] += support
msk_biser_support
```
