```python
dogs = ["mischka", 'mCanLor1.2', 'china', 'nala', 'tasha', 'wags', 'yella', 'zoey', 'sandy']
```

```python
dogs_to_assembly = {}
dogs_to_assembly['mischka'] = "UU_Cfam_GSD_1.0"
dogs_to_assembly['mCanLor1.2'] = "mCanLor1.2"
dogs_to_assembly['nala'] = 'Nala_ASM864105v3'
dogs_to_assembly['zoey'] = 'zoey.2.3'
dogs_to_assembly['yella'] = 'Yella_PRJNA610232'
dogs_to_assembly['china'] = 'China_UNSW_CanFamBas_1.2'
dogs_to_assembly['tasha'] = 'Dog10K_Boxer_Tasha_1.0.KP081776.1'
dogs_to_assembly['wags'] = 'Wags_PRJNA512874'
dogs_to_assembly['sandy'] = 'Sandy_ASM325472v2'
```

```python
gene_directory = {}
for dog in dogs:
    print(dog)
    gene_directory[dog] = {}
    f = open("/nfs/turbo/jmkiddscr/anthony-projects/gene_liftover/{}_gene_directory_PC.bed".format(dog), 'rt')
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        chro, start, end, gene = line[0], line[1], line[2], line[3]
        gene_directory[dog][gene] = (chro, start, end)
    f.close()
```

```python
for dog, genes in gene_directory.items():
    print(dog, len(genes))
```

```python
exon_directory = {}
exon_find_gene = {}
for dog in dogs:
    print(dog)
    exon_directory[dog] = {}
    exon_find_gene[dog] = {}
    f = open("/nfs/turbo/jmkiddscr/anthony-projects/gene_liftover/{}_gene_exon_directory_PC.bed".format(dog), 'rt')
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        chro, start, end = line[0], line[1], line[2]
        gene = line[3].split("_exon")[0]
        if gene not in exon_directory[dog]:
            exon_directory[dog][gene] = {}
        exon_num = line[3].split("_exon")[1]
        exon_directory[dog][gene][exon_num] = (chro, start, end)
        exon_find_gene[dog][(chro, start, end)] = gene
    f.close()
```

```python
for dog, genes in exon_directory.items():
    f = open("{}_exon_directory.txt".format(dog), 'w')
    for gene, exon_coords in genes.items():
        for num, coords in exon_coords.items():
            f.write('{}:{}-{}'.format(coords[0], coords[1], coords[2]) + '\n')
    f.close()
```

```python
cdna_seqs = {}
for dog in dogs:
    print(dog)
    cdna_seqs[dog] = {}
    f = open("{}.exon_seqs.fa".format(dog), 'rt')
    inFile = f.readlines()
    current_gene = ""
    current_cDNA = ""
    for line in inFile:
        if ">" in line:
            # get gene name
            line = line.lstrip(">").rstrip()
            chro = line.split(":")[0]
            start = line.split(":")[1].split("-")[0]
            end = line.split(":")[1].split("-")[1]
            gene = exon_find_gene[dog][(chro, start, end)]
            if current_gene == "":
                current_gene = gene
            if gene != current_gene:
                if current_gene not in cdna_seqs[dog].keys():
                    cdna_seqs[dog][current_gene] = ""
                cdna_seqs[dog][current_gene] += current_cDNA
                current_cDNA = ""
                current_gene = gene
        else:
            current_cDNA += line.rstrip()
    f.close()
```

```python
for dog, seqs in cdna_seqs.items():
    f = open("{d}_cDNA/{d}_all_cDNA.fa".format(d=dog), 'w')
    for gene, seq in seqs.items():
        f.write(">{}".format(gene) + '\n' + seq + '\n')
    f.close()
```

```python
num_list = []
for i in range(600):
    if len(str(i)) == 1:
        num = '00{}'.format(str(i))
    elif len(str(i)) == 2:
        num = '0{}'.format(str(i))
    else:
        num = str(i)
    num_list.append(num)
num_list
```

```python
f = open("mischka_blat_cDNA_commands.txt", 'w')
for num in num_list:
    f.write("blat /nfs/turbo/jmkidddata/genomes/UU_Cfam_GSD_1.0/ref/UU_Cfam_GSD_1.0.fa mischka_cDNA/mischka_cDNA_{n}.fa -ooc=/nfs/turbo/jmkidddata/genomes/UU_Cfam_GSD_1.0/ref/11.ooc blat_results/mischka_blat/mischka_{n}_blat_results.txt".format(n=num) + '\n')  
f.close()   
```

```python
for dog in dogs:
    if dog == "mischka":
        continue
    elif dog == "mCanLor1.2":
        print(dog)
        assembly = dogs_to_assembly[dog]
        f = open("{}_blat_cDNA_commands.txt".format(dog), 'w')
        for num in num_list:
            f.write("blat /nfs/turbo/jmkidddata/genomes/{a}/ref/{a}.fa {d}_cDNA/mcanlor_cDNA_{n}.fa -ooc=/nfs/turbo/jmkidddata/genomes/{a}/ref/11.ooc blat_results/{d}_blat/{d}_{n}_blat_results.txt".format(a=assembly,d=dog,n=num) + '\n')  
        f.close() 
    elif dog == "zoey":
        print(dog)
        assembly = dogs_to_assembly[dog]
        f = open("{}_blat_cDNA_commands.txt".format(dog), 'w')
        for num in num_list:
            f.write("blat /nfs/turbo/jmkidddata/genomes/{d}/assemblies/2.3/ref/{a}.fa {d}_cDNA/{d}_cDNA_{n}.fa -ooc=/nfs/turbo/jmkidddata/genomes/{d}/assemblies/2.3/ref/11.ooc blat_results/{d}_blat/{d}_{n}_blat_results.txt".format(a=assembly,d=dog,n=num) + '\n')  
        f.close()   
    else:
        print(dog)
        assembly = dogs_to_assembly[dog]
        f = open("{}_blat_cDNA_commands.txt".format(dog), 'w')
        for num in num_list:
            f.write("blat /nfs/turbo/jmkidddata/genomes/{a}/ref/{a}.fa {d}_cDNA/{d}_cDNA_{n}.fa -ooc=/nfs/turbo/jmkidddata/genomes/{a}/ref/11.ooc blat_results/{d}_blat/{d}_{n}_blat_results.txt".format(a=assembly,d=dog,n=num) + '\n')  
        f.close()   
```

```python
for dog in dogs:
    print(dog)
    f = open("{}_blast_cDNA_commands.txt".format(dog), 'w')
    for num in num_list:
        f.write("blastn -task dc-megablast -query {d}_cDNA/{d}_cDNA_{n}.fa  -db {d}.blast_rm -out blast_results/{d}_results/RESULTS-OUT-FILE-{n} -outfmt 7\n".format(d = dog, n = num))
    f.close()   
```
