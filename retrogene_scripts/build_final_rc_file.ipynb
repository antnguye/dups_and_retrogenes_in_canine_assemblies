```python
# read in fullORF
fullORF_set = {}
for dog in dogs:
    fullORF_set[dog] = []
    f = open("{}.fullORF_retrocopies.sorted.txt".format(dog), 'rt')
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        fullORF_set[dog].append(tuple(line[0:3]))
    f.close()
fullORF_set
```

```python
# check for segdup overlap
dog_retrocopies_in_fCN = {}
for dog in dogs:
    dog_retrocopies_in_fCN[dog] = []
    f = open("{}.retrocopy_intersect_fastCN.txt".format(dog), 'rt')
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        rc = (line[3], line[4], line[5])
        if rc not in dog_retrocopies_in_fCN[dog]:
            dog_retrocopies_in_fCN[dog].append(rc)
    f.close()
for dog, rcs in dog_retrocopies_in_fCN.items():
    print(dog, len(rcs))
```

```python
final_rc_file = {}
for dog in dogs:
    final_rc_file[dog] = {}
    f = open("{}.all_retrocopies_wUn.sorted.txt".format(dog), 'rt')
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        coord = (line[0], line[1], line[2])
        if coord in final_rc_file[dog].keys():
            print(dog, coord, line)
            print(final_rc_file[dog][coord])
            print()
        final_rc_file[dog][coord] = line # chro, start, end, gene, strand 
    f.close()
for dog, rcs in final_rc_file.items():
    print(dog, len(rcs))
```

```python
final_rc_file = {}
for dog in dogs:
    final_rc_file[dog] = {}
    f = open("{}.all_retrocopies_wUn.sorted.txt".format(dog), 'rt')
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        coord = (line[0], line[1], line[2])
        if coord in final_rc_file[dog].keys():
            final_rc_file[dog][coord] = [line[0], line[1], line[2], line[3] + '/' + final_rc_file[dog][coord][3], line[4]]
        else:
            final_rc_file[dog][coord] = line # chro, start, end, gene, strand 
    f.close()
for dog, rcs in final_rc_file.items():
    print(dog, len(rcs))
for dog in dogs:
    f = open("/nfs/turbo/jmkidddata/matt-projects/Stuff_For_Anthony/Identify_Hallmarks_From_Retrogenes/Scale_Up_2024_02_13/{}/Retrogene_Output_dist_70.txt".format(dog), 'rt')
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        chro, start, end, gene = line[0], line[1], line[2], line[4]
        retro = (chro, start, end)
        try:
            find_entry = final_rc_file[dog][retro]
        except: 
            print(dog, retro)
        hallmark_results = line[5:11]
        if len(find_entry) == 11:
            continue
        for item in hallmark_results:
            find_entry.append(item)
    f.close()
for dog, coord_to_list in final_rc_file.items():
    for coord, entry in coord_to_list.items():
        if len(entry) < 6:
            for i in range(6):
                entry.append("N/A")
for dog, coord_to_list in final_rc_file.items():
    for coord, entry in coord_to_list.items():
        if coord in fullORF_set[dog]:
            entry.append("True")
        else:
            entry.append("False")
for dog, coord_to_list in final_rc_file.items():
    for coord, entry in coord_to_list.items():
        if coord in dog_retrocopies_in_fCN[dog]:
            entry.append("True")
        else:
            entry.append("False")
#final_rc_file
```
