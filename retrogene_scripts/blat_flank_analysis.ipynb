```python
dog_to_assembly = {}
dog_to_assembly['china'] = "China_UNSW_CanFamBas_1.2"
dog_to_assembly['sandy'] = 'Sandy_ASM325472v2'
dog_to_assembly['nala'] = 'Nala_ASM864105v3'
dog_to_assembly['zoey'] = 'zoey'
dog_to_assembly['mischka'] = 'mischka'
```

```python
dog_autosomal_sizes = {}
for dog in dog_to_assembly.keys():
    #print(dog)
    f = open("../../fastCN/windows/segdup_analysis/{}.sorted.chrom.sizes".format(dog), 'rt')
    inFile = f.readlines()
    autosomal_size = 0
    chr_count = 0
    for line in inFile:
        line = line.rstrip().split()
        if "chrX" not in line[0] and "chrUn" not in line[0] and "chrY" not in line[0] and "chrM" not in line[0] and "CTG" not in line[0] and "scaffold" not in line[0]:
            autosomal_size += int(line[1])
            chr_count += 1
    dog_autosomal_sizes[dog] = autosomal_size
    print(dog, chr_count)
    f.close()
dog_autosomal_sizes
```

```python
dog_snps = {}
dog_snp_positions = {}
for dog, assembly in dog_to_assembly.items():
    print(dog)
    dog_snps[dog] = 0
    dog_snp_positions[dog] = []
    f = open("/nfs/turbo/jmkiddscr/anthony-projects/retrocopy_analysis/blast_results/get_snps/VCF/{}.vcf".format(assembly), 'rt')
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        if line[0][0] == "#":
            continue
        chro = line[0]
        pos = line[1]
        if "chrUn" in chro or "chrY" in chro or "chrM" in chro or "CTG" in chro or "scaffold" in chro or "chrX" in chro:
            continue
        if len(line[3]) == 1 and len(line[4]) == 1:
            dog_snps[dog] += 1
            dog_snp_positions[dog].append((chro, pos, pos))
    f.close()
dog_snps
```

```python
chrX_snps = {}
for dog, snps in dog_snp_positions.items():
    chrX_snps[dog] = 0
    for i in snps:
        if "chrX" in i[0]:
            chrX_snps[dog] += 1
chrX_snps
```

```python
dog_rcs_w_flanks = {}
dog_retrocopies = {}
for dog in dogs:
    dog_rcs_w_flanks[dog] = {}
    dog_retrocopies[dog] = {}
    f = open("{}.retrocopy_locations.txt".format(dog), 'rt')
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip()
        chro = line.split(":")[0]
        if "chrUn" in chro or "chrY" in chro or "chrM" in chro or "CTG" in chro or "scaffold" in chro or "chrX" in chro:
            continue
        new_start = int(line.split(":")[1].split("-")[0]) - 1000
        old_start = line.split(":")[1].split("-")[0]
        old_end = line.split("-")[1]
        new_end = int(line.split("-")[1]) + 1000
        dog_rcs_w_flanks[dog][line + "_5'"] = "{}:{}-{}".format(chro, new_start, old_start)
        dog_rcs_w_flanks[dog][line + "_3'"] = "{}:{}-{}".format(chro, old_end, new_end)

        if chro not in dog_retrocopies[dog].keys():
            dog_retrocopies[dog][chro] = []
        dog_retrocopies[dog][chro].append(old_start)
        dog_retrocopies[dog][chro].append(old_end)
    f.close()
dog_rcs_w_flanks
```

```python
dog_to_mcanlor = {}
for dog in dogs:
    print(dog)
    f = open("{}_to_mcanlor_flanks.paf".format(dog), 'rt')
    dog_to_mcanlor[dog] = {}
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        if int(line[11]) == 255:
            continue
        orig_coord = line[0]
        new_chro = line[5]
        if "chrX" in new_chro:
            continue
        new_start = line[7]
        new_end = line[8]
        new_coord = (new_chro, new_start, new_end)
        dog_to_mcanlor[dog][orig_coord] = new_coord
    f.close()
dog_to_mcanlor
```

```python
mcanlor_to_dog = {}
for dog in dogs:
    mcanlor_to_dog[dog] = {}
    f = open("mcanlor_to_{}_flanks.paf".format(dog), 'rt')
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        if int(line[11]) == 255:
            continue
        orig_coord = line[0]
        new_chro = line[5]
        if "chrX" in new_chro:
            continue
        new_start = line[7]
        new_end = line[8]
        new_coord = (new_chro, new_start, new_end)
        mcanlor_to_dog[dog][orig_coord] = new_coord
    f.close()
mcanlor_to_dog
```

```python
# step 7: determine counterparts (unique to mCanLor in each assembly)
```

```python
unique_to_mcl = {}
for dog, coords in mcanlor_to_dog.items():
    unplaced_matches = 0
    no_counterpart = 0
    mcl_counterpart = {}
    for mcl_coord, dog_coord in coords.items():
        chro = dog_coord[0]
        if "chrUn" in chro or "chrY" in chro or "chrM" in chro or "CTG" in chro or "scaffold" in chro or "chrX" in chro:
            unplaced_matches += 1
            continue
        start = int(dog_coord[1])
        end = int(dog_coord[2])
        dog_comparisons = dog_retrocopies[dog][chro]
        counterpart_found = False
        for coord in dog_comparisons:
            coord_up = int(coord)-1500
            coord_down = int(coord)+1500
            if start in range(coord_up,coord_down) or end in range(coord_up,coord_down):
                counterpart_found = True
                break
        if counterpart_found is True:
            if chro not in mcl_counterpart.keys():
                mcl_counterpart[chro] = 0
            mcl_counterpart[chro] += 1
        elif counterpart_found is False:
            no_counterpart += 1
    
    tot = 0
    for chro, counts in mcl_counterpart.items():
        tot += counts
    
    print(dog)
    print("Unplaced contig matches:", unplaced_matches)
    print("Avg retrocopies unique to MCL:", no_counterpart/2)
    print("Counterparts to {} found:".format(dog), tot/2)
    print("Number of assembled retrocopies in total:", len(mcanlor_to_dog[dog])/2)
    print()
    
    unique_to_mcl[dog] = no_counterpart/2
```
