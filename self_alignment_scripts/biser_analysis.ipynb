```python
dogs = ['mischka', 'mcanlor1.2', 'china', 'nala', 'sandy', 'tasha', 'wags', 'yella', 'zoey', 'UU_Cfam_GSD_1.0']
```

```python
all_biser = {}
all_biser_unplaced_partner = {}
for dog in dogs:
    print(dog)
    all_biser[dog] = []
    all_biser_unplaced_partner[dog] = []
    f = open("{d}/{d}-BISER.+unplaced.out.bed".format(d=dog), 'rt')
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        mismatch_rate = line[13].split(";")[0].split("=")[1]
        if float(mismatch_rate) > 10:
            continue
            
        dup1 = (line[0], line[1], line[2])
        dup2 = (line[3], line[4], line[5])
        
        if int(line[2]) - int(line[1]) > 1000:
            all_biser[dog].append(dup1)
        if int(line[5]) - int(line[4]) > 1000:
            all_biser[dog].append(dup2)
        
        if "chrUn" in dup1[0] and "chrUn" not in dup2[0]:
            all_biser_unplaced_partner[dog].append(dup2)
        elif "chrUn" not in dup1[0] and "chrUn" in dup2[0]:
            all_biser_unplaced_partner[dog].append(dup1)
```

```python
all_biser_merged = {}
for dog in dogs:
    print(dog)
    all_biser_merged[dog] = []
    f = open("{}.all_biser.merged.bed".format(dog), 'rt')
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        all_biser_merged[dog].append(tuple(line))
    f.close()
all_biser_merged
```

```python
for dog, dups in all_biser_merged.items():
    assembled_dup_count = 0
    for dup in dups:
        if "chrY" in dup[0] or "chrM" in dup[0] or "chrUn" in dup[0] or "CTG" in dup[0] or "scaffold" in dup[0]:
            if dog == "yella":
                print(dup)
        else:
            assembled_dup_count += 1
    print(dog, len(dups), assembled_dup_count, len(dups) - assembled_dup_count)
```

```python
# get dog sizes
dog_chrom_sizes = {}
for dog in dogs:
    dog_chrom_sizes[dog] = {}
    f = open("{}.sorted.chrom.sizes".format(dog), 'rt')
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        if "chrUn" in line[0] or "CTG" in line[0] or "scaffold" in line[0]:
            chro = "chrUn"
        else:
            chro = line[0]
        dog_chrom_sizes[dog][chro] = int(line[1])
dog_chrom_sizes
```

```python
mischka_auto_dups = []
for dog, dups in all_biser_merged.items():
    total_autosome_count = 0
    pre_1mb = 0
    pre_1mb_post_end = 0
    last_1mb = 0
    for dup in dups:
        if "chrX" in dup[0] or "chrY" in dup[0] or "chrM" in dup[0] or "chrUn" in dup[0] or "CTG" in dup[0] or "scaffold" in dup[0]:
            pass
        else:
            total_autosome_count += 1
            if int(dup[1]) < 1000000:
                pre_1mb += 1
            elif int(dup[1]) > (dog_chrom_sizes[dog][dup[0]] - 1000000):
                last_1mb += 1
                
            if dog == "mischka":
                mischka_auto_dups.append((dup[0],dup[1], dup[2]))
    print(dog, pre_1mb, round(pre_1mb/total_autosome_count * 100, 2), last_1mb, round(last_1mb/total_autosome_count * 100, 2), total_autosome_count)
```

```python
total_genome_sizes = {}
for dog in dogs:
    f = open("{}.sorted.chrom.sizes".format(dog), 'rt')
    total_genome_sizes[dog] = [0,0,0] # total, assembled, etc
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        if "chrY" in line[0] or "chrM" in line[0]:
            pass
        elif "chrUn" in line[0] or "CTG" in line[0] or "scaffold" in line[0]:
            total_genome_sizes[dog][2] += int(line[1])
        else:
            total_genome_sizes[dog][1] += int(line[1])
        total_genome_sizes[dog][0] += int(line[1])
    f.close()
total_genome_sizes
```

```python
total_bp = {}
for dog, dups in all_biser_merged.items():
    total_bp[dog] = [0,0,0] #total bp, assembled bp, etc. bp
    for dup in dups:
        size = int(dup[2]) - int(dup[1])
        if "chrY" in dup[0] or "chrM" in dup[0]:
            pass
        elif "chrUn" in dup[0] or "CTG" in dup[0] or "scaffold" in dup[0]:
            total_bp[dog][2] += size
        else:
            total_bp[dog][1] += size
        total_bp[dog][0] += size
    print(dog, [round(x/1000000,3) for x in total_bp[dog]])
    for i in range(len(total_bp[dog])):
        if dog == "yella":
            continue
        print(round(total_bp[dog][i] / total_genome_sizes[dog][i] * 100, 2))
```
