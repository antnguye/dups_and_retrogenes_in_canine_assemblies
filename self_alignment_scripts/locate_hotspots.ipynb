```python
def ucsc_form(coord_list):
    if int(coord_list[2]) > int(coord_list[1]):
        return("{}:{}-{}".format(coord_list[0], coord_list[1], coord_list[2]))
    else:
        return("{}:{}-{} -".format(coord_list[0], coord_list[2], coord_list[1]))
```

```python
f = open("UU_Cfam_GSD_1.0.sorted_biser.bed", 'rt')
sorted_biser = []
inFile = f.readlines()
for line in inFile:
    line = line.rstrip().split()
    sorted_biser.append(line)
f.close()
sorted_biser
```

```python
len(sorted_biser)
```

```python
hotspot_eligible = []
for item in sorted_biser:
    size = int(item[2]) - int(item[1])
    if size < 2500:
        hotspot_eligible.append(item)
hotspot_eligible
```

```python
len(hotspot_eligible)
```

```python
import math
```

```python
def round_up(n, decimals=0):
    multiplier = 10 ** decimals
    return math.ceil(n * multiplier) / multiplier
```

```python
def round_down(n, decimals=0):
    multiplier = 10 ** decimals
    return math.floor(n * multiplier) / multiplier
```

```python
def round_half_up(n, decimals=0):
    multiplier = 10 ** decimals
    return math.floor(n*multiplier + 0.5) / multiplier
```

```python
hotspot_to_area = {}
hotspot_members = {}
for item in hotspot_eligible:
    lower_b = round_down(int(item[1]), -2)
    upper_b = round_up(int(item[2]), -2)
    area = (item[0], str(int(lower_b)), str(int(upper_b)))
    if area not in hotspot_to_area.keys():
        hotspot_to_area[area] = 0
    hotspot_to_area[area] += 1
    if area not in hotspot_members.keys():
        hotspot_members[area] = []
    hotspot_members[area].append(area)
hotspot_to_area
```

```python
hotspot_members
```

```python
for item, v in hotspot_to_area.items():
    print(ucsc_form(item), v)
```

```python
f = open("mischka_hotspots.merged_counts.bed", 'rt')
inFile = f.readlines()
merged_hotspot_to_area = {}
for line in inFile:
    line = line.rstrip().split()
    hs = (line[0], line[1], line[2])
    if hs not in merged_hotspot_to_area.keys():
        merged_hotspot_to_area[hs] = 0
    orig_hs = (line[3], line[4], line[5])
    merged_hotspot_to_area[hs] += hotspot_to_area[orig_hs]
f.close()
merged_hotspot_to_area
```

```python
i = 0
for item, v in merged_hotspot_to_area.items():
    if v >= 4:
        i += 1
        print(ucsc_form(item), v)
```

```python
#how many hotspots, how many bp from a hotspot intersect?
unique_rcs = []
unique_hs = []
f = open("mischka.hotspot_rc_intersect.bed", 'rt')
inFile = f.readlines()
for line in inFile:
    line = line.rstrip().split()
    hotspot = (line[0], line[1], line[2])
    if hotspot not in unique_hs:
        unique_hs.append(hotspot)
    rc = (line[3], line[4], line[5], line[7])
    if rc not in unique_rcs:
        unique_rcs.append(rc)
f.close()
```
