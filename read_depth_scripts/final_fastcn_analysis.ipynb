```python
dogs = ["mischka","mCanLor1.2", "nala", "sandy", "tasha", "wags","yella","zoey"]
```

```python
all_fastcn = {}
all_fastcn_no_unplaced = {}
all_fastcn_10kb = {}
all_fastcn_10kb_no_unplaced = {}
for dog in dogs:
    all_fastcn[dog] = []
    all_fastcn_no_unplaced[dog] = []
    all_fastcn_10kb[dog] = []
    all_fastcn_10kb_no_unplaced[dog] = []
    if dog == "mischka":
        f = open("mischka_fastCN_segdups.bed", 'rt')
    else:
        f = open("{}_self.fastcn_segdups.bed".format(dog), 'rt')
    inFile = f.readlines()
    for line in inFile:
        line = line.rstrip().split()
        if dog == "mischka" and "chrY" in line[0]:
            continue
        all_fastcn[dog].append(line)
        if "chrUn" not in line[0] and "chrM" not in line[0] and "CTG" not in line[0] and "scaffold" not in line[0]:
            all_fastcn_no_unplaced[dog].append(line)
        size = int(line[2]) - int(line[1])
        if size > 10000:
            all_fastcn_10kb[dog].append(line)
        if "chrUn" not in line[0] and "chrM" not in line[0] and "CTG" not in line[0] and "scaffold" not in line[0] and size > 10000:
            all_fastcn_10kb_no_unplaced[dog].append(line)
    f.close()
all_fastcn
```

```python
e = open("mischka.sorted.chrom.sizes", 'rt')
chromFile = e.readlines()
chr_size_dict = {}
for line in chromFile:
    line = line.rstrip().split()
    chro = line[0]
    size = int(line[1])
    if "chrUn" in chro or "scaffold" in chro or "CTF" in chro:
        if "chrUn" not in chr_size_dict.keys():
            chr_size_dict["chrUn"] = size
        else:
            chr_size_dict["chrUn"] += size
    elif "chrY" in chro:
        if "chrY" not in chr_size_dict.keys():
            chr_size_dict["chrY"] = size
        else:
            chr_size_dict["chrY"] += size
    elif "chrM" in chro:
        if "chrM" not in chr_size_dict.keys():
            chr_size_dict["chrM"] = size
        else:
            chr_size_dict["chrM"] += size
    else:
        chr_size_dict[chro] = size

    if "total" not in chr_size_dict.keys():
        chr_size_dict["total"] = size
    else:
        chr_size_dict["total"] += size

    if "chrY" not in chr_size_dict.keys():
        chr_size_dict["chrY"] = 1
    if "chrUn" not in chr_size_dict.keys():
        chr_size_dict["chrUn"] = 1
    if "chrM" not in chr_size_dict.keys():
        chr_size_dict["chrM"] = 1
e.close()
```

```python
total_auto_size = 0
for chro, bp in chr_size_dict.items():
    if "total" in chro or "chrUn" in chro or "chrM" in chro:
        continue
    total_auto_size += bp
total_auto_size
```

```python
fmb = 0
lmb = 0
mischka_fastcn_10kb_no_unplaced_no1mb = []
for coord in all_fastcn_10kb_no_unplaced['mischka']:
    if "chrX" in coord[0]:
        mischka_fastcn_10kb_no_unplaced_no1mb.append(coord)
    elif int(coord[1]) < 1000000:
        fmb += 1
    else:
        mischka_fastcn_10kb_no_unplaced_no1mb.append(coord)
    if "chrX" not in coord[0] and int(coord[2]) > (chr_size_dict[coord[0]] - 1000000):
        lmb += 1
fmb, lmb
```
